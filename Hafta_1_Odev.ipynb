Modeli Eğitme Kısmı:

```python
features = df[['island', 'bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'year']] # --> Bağımsız değişkenler
label = df['species'] # --> Bağımlı değişken

display(features.head())
display(label.head())
```

Eğitim Veriseti ve Test Verisetinin Belirlenmesi:

```python
from sklearn.model_selection import train_test_split
# x: bağımsız değişkenler
# y: bağımlı değişken

x_train, x_test, y_train, y_test = train_test_split(features, label, test_size=0.2, random_state=42)

# x_train bağımsız değişken eğitim verileri %80
# x_test bağımsız değişken test verileri %20
# y_train bağımlı değişken eğitim veri seti %80
# y_test bağımlı değişken test veri seti %20

print("\nEğitim ve Test Seti Boyutları:")
print("X_train:", x_train.shape)
print("X_test:", x_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```
