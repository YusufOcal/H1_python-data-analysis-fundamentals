## EKSİK DEĞER

Elbette, eksik değer oranı yüksek olan bir veri setinde farklı doldurma stratejilerini kodlarla birlikte açıklıyorum.

### 1. **Eksik Değeri Yüksek Sütunları Kaldırma**

Eğer bir sütunda çok fazla eksik değer varsa ve bu sütun analiz için kritik değilse, sütunu tamamen kaldırabilirsiniz.

```python
# Eksik değeri %50'den fazla olan sütunları kaldırmak
threshold = 0.5 * len(features)  # Örneğin, veri setinin %50'sinden fazla eksik olan sütunları belirleme
features = features.dropna(thresh=threshold, axis=1)
print(features.head())
```

Bu kod, %50'den fazla eksik değer içeren sütunları kaldırır.

### 2. **Eksik Değeri Yüksek Satırları Kaldırma**

Belirli satırlarda birçok eksik değer varsa, bu satırları veri setinden çıkarabilirsiniz.

```python
# Eksik değeri olan satırları kaldırmak
features = features.dropna(how='any')  # En az bir eksik değeri olan satırları kaldırır
print(features.head())
```

Bu kod, herhangi bir sütunda eksik veri içeren tüm satırları kaldırır.

### 3. **En Sık Görülen Değer (Mod) ile Doldurma**

Kategorik sütunlarda eksik veriyi doldurmak için en sık görülen değeri kullanabilirsiniz.

```python
# Cabin sütunundaki eksik değerleri en sık görülen değer (mod) ile doldurma
features['Cabin'] = features['Cabin'].fillna(features['Cabin'].mode()[0])
print(features['Cabin'].head())
```

Bu kod, `Cabin` sütunundaki eksik değerleri en sık görülen değerle doldurur.

### 4. **Sabit Bir Değer ile Doldurma**

Eğer eksik değerin kendisini özel bir durum olarak ifade etmek istiyorsanız, eksik değerleri sabit bir değerle doldurabilirsiniz.

```python
# Cabin sütunundaki eksik değerleri 'Unknown' olarak doldurma
features['Cabin'] = features['Cabin'].fillna('Unknown')
print(features['Cabin'].head())
```

Bu kod, `Cabin` sütunundaki eksik değerleri `'Unknown'` olarak doldurur.

### 5. **K-NN İmputation (K En Yakın Komşu ile Doldurma)**

`KNNImputer`, eksik değerleri komşu değerlerle doldurur ve sayısal verilerde kullanılır. `KNeighborsClassifier` sınıfı kullanılarak uygulanabilir.

```python
from sklearn.impute import KNNImputer

# KNNImputer kullanarak eksik değerleri doldurma (K = 5)
imputer = KNNImputer(n_neighbors=5)
features[['Age', 'Fare']] = imputer.fit_transform(features[['Age', 'Fare']])
print(features[['Age', 'Fare']].head())
```

Bu kod, `Age` ve `Fare` sütunlarındaki eksik değerleri komşu değerlerle doldurur. `n_neighbors=5` parametresi, 5 en yakın komşuyu kullanarak doldurulacağını belirtir.

### 6. **Makine Öğrenmesi Tabanlı Doldurma (Örneğin, Regresyon ile Tahmin)**

Eksik değerleri tahmin etmek için bir regresyon modeli kullanabilirsiniz. Eksik değerlerin olduğu sütunu bağımlı değişken olarak seçip, diğer sütunları bağımsız değişken olarak kullanabilirsiniz.

Örneğin, `Age` sütunundaki eksik değerleri diğer değişkenlerle tahmin etmek için bir basit regresyon modeli kurabilirsiniz:

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Age sütunundaki eksik olmayan satırları ayırıyoruz
train_data = features[features['Age'].notnull()]
test_data = features[features['Age'].isnull()]

# Bağımsız değişkenler (diğer sütunlar) ve bağımlı değişken (Age) olarak ayırma
X_train = train_data.drop(columns=['Age'])
y_train = train_data['Age']
X_test = test_data.drop(columns=['Age'])

# Basit bir lineer regresyon modeli oluşturup eğitiyoruz
model = LinearRegression()
model.fit(X_train, y_train)

# Eksik Age değerlerini tahmin ediyoruz ve dolduruyoruz
features.loc[features['Age'].isnull(), 'Age'] = model.predict(X_test)
print(features['Age'].head(10))
```

Bu kod, `Age` sütunundaki eksik değerleri diğer değişkenlerle tahmin eder ve doldurur.

### 7. **Eksik Değer İçeren Özelliklerin Özel İşaretlenmesi**

Eksik değerleri bir bilgi olarak korumak istiyorsanız, eksik olup olmadığını işaretleyebilirsiniz. Bu, özellikle eksikliğin kendisinin anlam taşıdığı durumlarda faydalıdır.

```python
# Cabin bilgisinin eksik olup olmadığını belirten bir sütun ekleme
features['Cabin_Missing'] = features['Cabin'].isnull().astype(int)
print(features[['Cabin', 'Cabin_Missing']].head())
```

Bu kod, `Cabin` sütunundaki eksik veriyi `1` (eksik) veya `0` (eksik değil) şeklinde işaretleyen yeni bir sütun ekler.

## EKSİK DEĞERLERDEN NE ZAMAN KURTULMAM GEREKLİ?

### 1. **Eksik Değeri Çok Olan Sütunları Kaldırma**

Eğer bir sütunda çok fazla eksik değer varsa ve bu sütun analiz için kritik değilse, o sütunu veri setinden tamamen kaldırabilirsiniz. Aşağıda, bir sütundaki eksik veri oranı %50’den fazlaysa bu sütunun kaldırılmasını gösteren bir örnek verilmiştir.

```python
# Kaldırma eşiğini belirliyoruz (%50 eksik veri oranı)
threshold = 0.5 * len(features)  # Toplam satır sayısının %50'si

# Eksik değeri %50'den fazla olan sütunları kaldırma
features = features.dropna(thresh=threshold, axis=1)
print("Eksik değeri %50'den fazla olan sütunlar kaldırıldı:")
print(features.head())
```

Bu kodda:
- `thresh=threshold` parametresi, sütunun en az belirli bir sayıda (bu örnekte toplam satır sayısının %50'si) dolu değer içermesi gerektiğini belirtir.
- `axis=1` ile sütun bazında işlem yapılmasını sağlarız.
  
Bu işlem sonunda, eksik değerin çok fazla olduğu sütunlar veri setinden çıkarılmış olur.

### 2. **Eksik Değeri Çok Olan Satırları Kaldırma**

Eğer belirli satırlarda çok sayıda eksik değer varsa, bu satırları veri setinden çıkarabilirsiniz. Bu durumda, satırda herhangi bir eksik değer varsa veya satırdaki belirli sayıda değerin eksik olduğu durumları kaldırabilirsiniz.

#### Tüm Eksik Değerleri İçeren Satırları Kaldırma

```python
# En az bir eksik değeri olan tüm satırları kaldırma
features = features.dropna(how='any')
print("En az bir eksik değeri olan satırlar kaldırıldı:")
print(features.head())
```

Bu kod, veri setinde herhangi bir sütunda eksik veri içeren tüm satırları kaldırır. `how='any'` parametresi, en az bir eksik değere sahip olan satırları hedef alır.

#### Tüm Değerleri Eksik Olan Satırları Kaldırma

```python
# Sadece tüm değerleri eksik olan satırları kaldırma
features = features.dropna(how='all')
print("Tüm değerleri eksik olan satırlar kaldırıldı:")
print(features.head())
```

Bu kod, sadece tüm sütunlardaki değerlerin eksik olduğu satırları kaldırır. `how='all'` parametresi, yalnızca tüm değerleri eksik olan satırları hedef alır.

#### Belirli Sayıda Eksik Değere Göre Satırları Kaldırma

Örneğin, bir satırda 3 veya daha fazla eksik değer varsa bu satırları kaldırmak isteyebilirsiniz.

```python
# En az 3 eksik değeri olan satırları kaldırma
features = features.dropna(thresh=len(features.columns) - 3)
print("En az 3 eksik değeri olan satırlar kaldırıldı:")
print(features.head())
```

Bu kodda:
- `thresh=len(features.columns) - 3` parametresi, her satırda en az (toplam sütun sayısı - 3) kadar dolu değere sahip olma şartını koyar. Yani, 3 veya daha fazla eksik değeri olan satırları kaldırır.

---

Bu yöntemler eksik veri oranı çok yüksek olan durumlarda veri setini temizlemek için kullanılabilir. Ancak, satır veya sütunları kaldırmak, veri kaybına yol açabilir; bu yüzden bu yöntemleri kullanırken dikkatli olmak önemlidir.

## VERİYİ EĞİTİM VE TEST OLARAK AYIRMA
Veriyi eğitim ve test olarak ikiye bölmek, makine öğrenmesi projelerinde başarılı bir model geliştirmek için temel bir adımdır. Bu işlemin amacı, modelimizin sadece mevcut veriyi "ezberlemekten" ziyade, yeni veriler üzerinde de doğru tahminler yapabilecek kadar "öğrenmesini" sağlamaktır.

### Mantığını Anlayalım

Bir insanın yeni bir konuyu öğrenme sürecini düşünelim. Örneğin, bir öğrenci matematik çalışırken önce ders kitabındaki konuları ve örnekleri inceleyerek öğrenir. Bu süreçte öğrendiği bilgileri uygulamak için soruları çözer. Bu aslında bir "eğitim" sürecidir. Ancak, öğrencinin gerçekten öğrendiğini ve bu bilgiyi yeni sorularda da uygulayabildiğini görmek için, sınavda daha önce karşılaşmadığı sorular sorulur. Sınav, öğrenci için bir testtir ve öğrendiklerini sınamak içindir. Eğer öğrenci sadece ders kitabındaki örnekleri ezberlemişse, sınavdaki farklı sorular karşısında zorlanabilir. Ama konuyu gerçekten anladıysa, sınavda başarılı olur.

Aynı mantıkla, bir makine öğrenmesi modelini de iki aşamada değerlendiririz:
1. **Eğitim aşaması**: Modeli veri üzerinde eğiterek ona "öğrenmesi" için bir fırsat tanırız.
2. **Test aşaması**: Modelin daha önce görmediği veri üzerinde ne kadar doğru tahmin yapabildiğini ölçeriz.

Bu bölme işlemi sayesinde, modelin "ezberleme" yapmadan, verinin genel özelliklerini öğrenip öğrenmediğini değerlendirebiliriz. Yani, modelin "genelleme yapabilme" yeteneğini test etmiş oluruz.

### Neden Bölüyoruz?

Bir modeli değerlendirmek için eğitim verisi yeterli değil çünkü bu veriyle modeli eğitirken, model bu veriyi zaten görmüş olur. Eğer modelimizi sadece eğitim verisiyle test edersek, sonuçlar yanıltıcı olabilir. Model, eğitim verisindeki detaylara aşina olduğu için bu veri üzerinde çok iyi performans gösterebilir. Ama aynı model, gerçek hayatta yeni verilerle karşılaştığında başarısız olabilir. Bu yüzden, modelin daha önce görmediği veriyle (yani test verisiyle) değerlendirilmesi gerekir.

### Özetle

Veriyi eğitim ve test olarak ikiye bölmek:
- Modelin ezberlemesini değil, öğrenmesini sağlamaya çalışır.
- Modelin gerçek dünyada yeni verilerle karşılaştığında da başarılı olup olmadığını değerlendirmemizi sağlar.

Bu bölme işlemi, bir insanın öğrenme sürecindeki pratik yapma ve sınav aşamalarına benzer. Eğitim veri seti, modelin öğrenmesi için kullanılırken, test veri seti modelin performansını değerlendirmek için kullanılır. Bu şekilde, modelin gerçekten öğrendiğini ve yeni veriler üzerinde doğru tahmin yapma yeteneğini ölçebiliriz.

````python

# Veri Setini Eğitim ve Test Olarak Ayırma
from sklearn.model_selection import train_test_split
#x: bağımsız değişkenler
#y : bağımlı değişken
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.2, random_state=42)

print("\nEğitim ve Test Seti Boyutları:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
````

Bu satır, makine öğrenmesi modelleri için veriyi eğitim ve test seti olarak ayırmak amacıyla kullanılan `train_test_split` fonksiyonunu gösterir. Bu fonksiyon, veriyi belirli bir oranda rastgele böler ve modelin doğruluğunu değerlendirmek için kullanılır. Bu işlemi parçalara ayırarak açıklayalım:

### Kod Açıklaması

```python
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.2, random_state=42)
```

Bu kod, `features` ve `label` veri setlerini ikiye böler: Eğitim seti ve test seti.

### Parametrelerin Anlamı

1. **`features`**:
   - Modelin öğrenmesi için kullanılan bağımsız değişkenlerin (özelliklerin) bulunduğu veri setidir. Bu, giriş verilerini (input) temsil eder.
   - `features`, modelin tahmin yapması için ihtiyaç duyduğu verilerden oluşur. Örneğin, bir tahmin modelinde yaş, gelir, cinsiyet gibi değişkenler `features` olarak kullanılır.

2. **`label`**:
   - Modelin tahmin etmeye çalıştığı bağımlı değişkendir. Bu, hedef değişkeni (output) temsil eder.
   - `label`, modelin tahmin etmesi istenen değerleri içerir. Örneğin, bir fiyat tahmin modelinde tahmin edilmek istenen fiyat değeri `label` olur.

3. **`test_size=0.2`**:
   - Verinin ne kadarının test setine ayrılacağını belirler. `test_size=0.2`, veri setinin %20'sinin test setine, %80'inin eğitim setine ayrılacağı anlamına gelir.
   - Bu parametre, test setinin büyüklüğünü ayarlamak için kullanılır. Örneğin, `test_size=0.3` olarak ayarlanırsa, veri setinin %30'u test seti olarak ayrılır.

4. **`random_state=42`**:
   - Veriyi bölme işleminin rastgeleliğini kontrol etmek için kullanılan bir parametredir.
   - `random_state` aynı değere ayarlandığında, her çalıştırmada aynı şekilde veri bölünür. Bu, sonuçların tekrarlanabilir olmasını sağlar. Farklı bir sayı değeri verilirse veya tamamen kaldırılırsa, verinin bölünmesi her seferinde farklı olur.
   - `random_state=42` ifadesi, sıklıkla kullanılan bir "rastgele durum" örneğidir; herhangi bir özel anlamı yoktur, sadece belirli bir rastgeleliği garanti eder.

### Döndürülen Değerler

- **`X_train`**:
  - `features` veri setinin eğitim setine ayrılan kısmıdır.
  - Modelin eğitilmesi (fit edilmesi) için kullanılan giriş verilerini (özellikleri) içerir.

- **`X_test`**:
  - `features` veri setinin test setine ayrılan kısmıdır.
  - Model eğitildikten sonra, modelin performansını değerlendirmek için kullanılan giriş verilerini içerir.

- **`y_train`**:
  - `label` veri setinin eğitim setine ayrılan kısmıdır.
  - Modelin öğrenmesi gereken hedef değişkenin değerlerini içerir.

- **`y_test`**:
  - `label` veri setinin test setine ayrılan kısmıdır.
  - Model eğitildikten sonra, modelin tahminlerini gerçek değerlerle karşılaştırmak için kullanılan hedef değişkenin değerlerini içerir.

### Örnek Açıklama

Örneğin, elimizde bir ev fiyat tahmin veri seti olduğunu düşünelim:
- **`features`** sütunları: Evin büyüklüğü, oda sayısı, bulunduğu bölge.
- **`label`** sütunu: Evin satış fiyatı.

Bu kod, veri setini %80 eğitim ve %20 test seti olarak ayırır:
- `X_train` ve `y_train` modelin eğitilmesi için kullanılır.
- `X_test` ve `y_test`, model eğitildikten sonra performansını test etmek için kullanılır.

### Neden Veri Setini Ayırıyoruz?

Veri setini ayırmamızın amacı, modelin performansını objektif bir şekilde değerlendirmektir. Modelin önce `X_train` ve `y_train` verileri üzerinde eğitilmesi sağlanır. Ardından `X_test` ve `y_test` verileri kullanılarak, modelin gerçek veriler üzerinde nasıl tahminler yaptığına bakılır. Bu işlem, modelin gerçek dünyada yeni verilere nasıl yanıt vereceğini simüle etmeye yardımcı olur.

Bu ayrım, modelin sadece eğitim verisine ezber yapmasını (overfitting) önler ve modelin genelleme yeteneğini ölçmemizi sağlar.

## ÖDEV DATASET HİKAYESİ
**Palmer Archipelago Penguins Dataset** (Palmer Penguen Veri Seti), Palmer Archipelago bölgesindeki üç farklı penguen türü hakkında biyometrik veriler içeren bir veri setidir. Bu veri seti, penguenlerin türlerini tanımlamak için kullanılır ve özellikle makine öğrenmesi, veri analizi ve görselleştirme eğitimleri için popülerdir. **Palmer Archipelago** ise, Antarktika'nın kuzey ucunda bulunan bir takımadadır ve burada birçok penguen kolonisi bulunur.

Veri seti, Palmer Station (Palmer İstasyonu) bölgesinde yaşayan üç penguen türünden toplanan verilerden oluşur: **Adelie**, **Chinstrap** ve **Gentoo**. Bu üç tür, görünüş ve boyut olarak birbirinden farklıdır ve veri seti, bu türler arasındaki farkları analiz etmek için kullanılır.

### Veri Setinin Hikayesi ve Amacı

Palmer Archipelago Penguins Dataset, farklı penguen türlerinin fiziksel özelliklerini kullanarak bu türleri sınıflandırmayı amaçlar. Penguenlerin türleri, yaşadıkları ortam ve boyutları hakkında bilgi vererek araştırmacıların bu canlıları daha iyi tanımasını sağlar. Bu veri seti, özellikle bilim insanlarının ve öğrencilerin, biyolojik türler arasındaki farklılıkları anlamaları ve makine öğrenmesi gibi tekniklerle veri analizi becerilerini geliştirmeleri için idealdir.

Veri seti, genellikle **Iris veri setine** bir alternatif olarak sunulmaktadır, çünkü daha fazla biyolojik çeşitlilik ve daha karmaşık analizlere olanak tanır. Aynı zamanda görselleştirme açısından da çeşitli ilişkileri incelemek için idealdir.

### Veri Setindeki Özellikler (Feature'lar)

Veri setinde her bir penguen için aşağıdaki özellikler bulunmaktadır:

1. **species**:
   - **Açıklama**: Penguenin türü. Bu hedef değişken olup, sınıflandırma için kullanılır.
   - **Değerler**: `'Adelie'`, `'Chinstrap'`, `'Gentoo'`
   - **Tip**: Kategorik

2. **island**:
   - **Açıklama**: Penguenin yaşadığı ada. Bu özellik, penguenlerin farklı coğrafi konumlarına göre değişiklik gösterebilir.
   - **Değerler**: `'Biscoe'`, `'Dream'`, `'Torgersen'`
   - **Tip**: Kategorik

3. **bill_length_mm**:
   - **Açıklama**: Penguenin gaga uzunluğu (milimetre cinsinden). Gaga uzunluğu, türler arasında değişiklik gösterebileceğinden sınıflandırma için önemli bir özellik olabilir.
   - **Tip**: Sayısal (float)

4. **bill_depth_mm**:
   - **Açıklama**: Penguenin gaga derinliği (milimetre cinsinden). Gaga derinliği de penguen türlerinin tanımlanmasında yardımcı olabilir.
   - **Tip**: Sayısal (float)

5. **flipper_length_mm**:
   - **Açıklama**: Penguenin yüzgeç uzunluğu (milimetre cinsinden). Yüzgeç uzunluğu, penguen türleri arasında farklılık gösterebilir.
   - **Tip**: Sayısal (float)

6. **body_mass_g**:
   - **Açıklama**: Penguenin vücut kütlesi (gram cinsinden). Türler arasında vücut kütlesi önemli bir ayırt edici özellik olabilir.
   - **Tip**: Sayısal (float)

7. **sex**:
   - **Açıklama**: Penguenin cinsiyeti. Cinsiyet, penguenlerin bazı özelliklerini etkileyebilir, ancak tüm veri noktalarında cinsiyet bilgisi bulunmamaktadır.
   - **Değerler**: `'Male'`, `'Female'`
   - **Tip**: Kategorik

8. **year**:
   - **Açıklama**: Verinin toplandığı yıl. Bu özellik, yıllar arasındaki farklılıkları incelemek için kullanılabilir.
   - **Tip**: Sayısal (integer)

### Veri Setindeki Hedef (Target) Değişken: LABEL:Bağımlı Değişken

Bu veri setinde **hedef değişken (label)** `species` sütunudur. Bu özellik, her bir penguenin türünü belirtir ve sınıflandırma modelleriyle bu türleri tahmin etmek için kullanılır.
